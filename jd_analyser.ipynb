## Importing Libraries

```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
import nltk
import re
```

## Data Preprocessing and Word Cloud Generation

```python
# Input text containing words

text = "Preferred Qualifications:\n Graduating December 2024 onwards with intent to return to degree-program after the completion of the internship.\n Experience in Kubernetes-related Cloud Native development, such as Kubernetes API, Operator, Custom Resource Definition (CRD) development, or other related Open Source projects. \n Experience with backend development and frameworks such as Gin, Iris, Django, Spring, etc. \n Deep understanding of data structure design, networking, authentication and data security. \n Familiarity with the whole web stack, including protocols and web server optimization techniques. \n Experience in using and contributing to Cloud Native projects, e.g. Prometheus , Istio/Linkerd, Envoy, KubeEdge, ArgoCD, OPA, Harbor, etc."
```

```python
# custom stopwords

custom_stopwords = ["qualifications", "experience", "such", "as", "etc", 'experience', 'development', 'related']
```

```python
# Preprocess the text by converting it to lowercase
text = text.lower()
```

```python
# regular expressions to split the text into words
words = re.findall(r'\b\w+\b', text)
```

```python
# Filtering out custom stopwords
filtered_words = [word for word in words if word not in custom_stopwords]

# Joining the filtered words back into a single string
filtered_text = ' '.join(filtered_words)
```

```python
filtered_text
```

```python
# Generating a word cloud
wc = WordCloud(
    background_color='white',
    width=2000, height=1000
).generate(filtered_text)

# Displaying the word cloud
plt.figure(figsize=(10, 5))  # Adjust the figure size for better visibility
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')  # Remove axes

# Saving the word cloud as an image
plt.savefig("word_cloud.png")

# displaying wordcloud
plt.show()
```

```python
# Extract the words from the word cloud
word_frequencies = wc.words_

# Convert the words to a list
words_list = list(word_frequencies.keys())

# Print the list of words
print(words_list)
```
